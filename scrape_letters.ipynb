```python
import time
import requests
import json
from pathlib import Path
from tqdm.notebook import tqdm
```

```python
register_request = "https://correspsearch.net/api/v1.2/tei-json.xql?sender=http://d-nb.info/gnd/118607960"
letter_request_template = "https://august-wilhelm-schlegel.de/briefedigital/letters/xml/{}"
```

```python
r = requests.get(register_request)
r_parsed = json.loads(r.content.decode("utf-8"))
```

```python
out_dir = Path("out/")
out_dir.mkdir(exist_ok=True)

for item in tqdm(r_parsed['teiHeader']['profileDesc']['correspDesc']):
    if 'key' in item and "august-wilhelm-schlegel" in item.get("ref", ""):
        corresp_key = item['key']
        request_key = corresp_key
        html_response = requests.get(item.get("ref"))
        if html_response.status_code == 200:
            if corresp_key not in html_response.url:
                schlegel_key = html_response.url.split("view/")[1].split("?")[0]
                print(f"{corresp_key} mapped to {schlegel_key}")
                request_key = schlegel_key
                
        xml_response = requests.get(letter_request_template.format(request_key))
        
        if xml_response.status_code == 200:
            with open(out_dir / f'{request_key}.xml', "wb") as fout:
                fout.write(xml_response.content)
                time.sleep(.1)
```
